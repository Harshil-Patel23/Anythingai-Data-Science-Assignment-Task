```python
total_closed_pnl=historical_data['Closed PnL'].size
```

```python
total_profitable_trades = historical_data[historical_data['Closed PnL'] > 0].shape[0]
```

```python
percentage_profitable_trades = (total_profitable_trades / total_closed_pnl) * 100
percentage_profitable_trades  
```

```python
# 1. Count trades per account
trade_counts = historical_data['Account'].value_counts()

# 2. Total trades
total_trades = len(historical_data)

# 3. Top 5 accounts total
top5_trades = trade_counts.head(5).sum()

# 4. Percentage contribution
top5_percentage = (top5_trades / total_trades) * 100

top5_percentage
```

```python
historical_data.head()
```

```python
# Group by each account
group = historical_data.groupby('Account')

account_stats = pd.DataFrame({
    'total_trades': group.size(),
    'profitable_trades': group.apply(lambda x: (x['Closed PnL'] > 0).sum()),
    'win_rate': group.apply(lambda x: (x['Closed PnL'] > 0).mean() * 100),
    'total_pnl': group['Closed PnL'].sum(),
    'avg_pnl': group['Closed PnL'].mean(),
    'avg_position_size': group['Size USD'].mean(),
    'long_percentage': group.apply(lambda x: (x['Side'] == 'Buy').mean() * 100),
})

account_stats
```

```python
# How many trades each coin has
coin_counts = historical_data['Coin'].value_counts()
coin_counts.head(20)
```

```python
total_trades = len(historical_data)
top10_trades = coin_counts.head(10).sum()

top10_percentage = (top10_trades / total_trades) * 100
top10_percentage
```

```python
coin_volume = historical_data.groupby('Coin')['Size USD'].sum().sort_values(ascending=False)

top10_volume = coin_volume.head(10).sum()
total_volume = coin_volume.sum()

top10_volume_percentage = (top10_volume / total_volume) * 100
top10_volume_percentage
```

```python
win_rate_per_account = historical_data.groupby('Account')['Closed PnL'].apply(lambda x: (x > 0).mean() * 100)
win_rate_per_account
```
